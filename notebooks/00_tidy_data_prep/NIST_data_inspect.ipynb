# Get particle classifications

```python
classes = {}
elements_list = []

with open(path + 'mllsq.hdz', 'r', encoding='utf-8', errors='replace') as f:
    for line in f:
        line = line.strip()
        match = re.match(r'CLASS(\d+)=(.*)', line)
        if match:
            class_id = int(match.group(1))
            class_name = match.group(2)
            classes[class_id] = class_name
        # Parse element definitions
        match = re.match(r'ELEM(\d+)=(\w+)\s+(\d+)\s+(\d+)', line)
        if match:
            elements_list.append(match.group(2))

print(f"Elements measured: {elements_list}")
print(f"\nNumber of classes: {len(classes)}")
print(f"\nAll class definitions:")
for k, v in sorted(classes.items()):
    print(f"  CLASS{k:2d} = {v}")
```

```python
mllsq = pd.read_csv(path + 'MLLSQ_maxParticle.csv', sep='\t', header=None, names=['particle_start', 'particle_end', 'classification'])
```

```python
def extract_class(s):
    match = re.search(r'\[(.+?):', s)
    if match:
        return match.group(1)
    return 'Unknown'

mllsq['class_name'] = mllsq['classification'].apply(extract_class)
```

```python
rows = []
for _, row in mllsq.iterrows():
    for pid in range(int(row['particle_start']), int(row['particle_end']) + 1):
        rows.append({
            'particle_id': pid,
            'class_name': row['class_name']
        })

particles_classified = pd.DataFrame(rows)
print(f"\nTotal individual particles: {len(particles_classified)}")
print(f"\nParticle class distribution:")
print(particles_classified['class_name'].value_counts().head(20))
```

```python
print("\nClasses containing GSR-relevant terms:")
gsr_keywords = ['GSR', 'Pb', 'Sb', 'Ba', 'Lead', 'Antimony', 'Barium']
for class_id, class_name in sorted(classes.items()):
    if any(kw.lower() in class_name.lower() for kw in gsr_keywords):
        print(f"  CLASS{class_id} = {class_name}")
```

```python
def label_gsr(class_name):
    class_lower = class_name.lower()
    if 'classic gsr' in class_lower or 'gsr' in class_lower:
        return 'GSR'
    elif class_lower in ['#unclassified#', 'lowcounts', 'low counts']:
        return 'Low_Quality'
    else:
        return 'Non_GSR'

particles_classified['gsr_label'] = particles_classified['class_name'].apply(label_gsr)
print("\nGSR label distribution (Shooter #1):")
print(particles_classified['gsr_label'].value_counts())
print(f"\nDetailed class distribution:")
print(pd.crosstab(particles_classified['class_name'], particles_classified['gsr_label']))
```
